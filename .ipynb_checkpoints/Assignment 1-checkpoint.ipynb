```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'smartphones.csv'
data = pd.read_csv(file_path)

# Handle missing values by filling with median
ram_median = data['RAM'].median()
storage_median = data['Storage'].median()
data['RAM'] = data['RAM'].fillna(ram_median)
data['Storage'] = data['Storage'].fillna(storage_median)

# Check for missing values
print("\nMissing values after handling:")
print(data.isnull().sum())

# Statistical summary
stats = data.describe().loc[['mean', 'std', '50%']]
print("\nStatistical Summary:")
print(stats)

# Function to plot PDF and CDF
def plot_pdf_cdf(data, label, color):
    plt.figure(figsize=(12, 5))

    # PDF
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True, color=color, stat="density")
    plt.title(f'Probability Density Function of {label}')
    plt.xlabel(label)
    plt.ylabel('Density')
    plt.grid(True)

    # CDF
    plt.subplot(1, 2, 2)
    data_sorted = np.sort(data)
    p = 1. * np.arange(len(data)) / (len(data) - 1)
    plt.plot(data_sorted, p, color=color)
    plt.title(f'Cumulative Distribution Function of {label}')
    plt.xlabel(label)
    plt.ylabel('CDF')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Plot PDF, CDF, and histogram for RAM, Storage, and Price
plot_pdf_cdf(data['RAM'], 'RAM', 'blue')
plot_pdf_cdf(data['Storage'], 'Storage', 'green')
plot_pdf_cdf(data['Price'], 'Price', 'red')

# Calculate and print correlation coefficients
correlation_ram_price = data['Price'].corr(data['RAM'])
correlation_storage_price = data['Price'].corr(data['Storage'])
print(f"\nCorrelation between Price and RAM: {correlation_ram_price:.2f}")
print(f"Correlation between Price and Storage: {correlation_storage_price:.2f}")

# Display scatter plots to visualize relationships
plt.figure(figsize=(10, 5))

# Scatter plot Price vs RAM
plt.subplot(1, 2, 1)
sns.scatterplot(x=data['RAM'], y=data['Price'])
plt.title('Price vs RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Price ($)')
plt.grid(True)

# Scatter plot Price vs Storage
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['Storage'], y=data['Price'])
plt.title('Price vs Storage')
plt.xlabel('Storage (GB)')
plt.ylabel('Price ($)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
